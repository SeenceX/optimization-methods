<h2>Численные методы многомерной оптимизации (методы первого порядка)</h2>

```python
from math import sqrt

def f(c, vector):
    return c[0]*vector[0]**2 + c[1]*vector[0]*vector[1] + c[2]*vector[1]**2


def derivF(x, y, k):
    #print(f"Градиент функции в точке x_{k} ({x}, {y}) = {[2*c[0] * x + c[1]*y, c[1]*x + 2*c[2]*y]}")
    return [2*c[0] * x + c[1]*y, c[1]*x + c[2]*2*y]


def norma(x, y, grad, k):
    print(f"Вычислим норму вектора ||{grad}||")
    return sqrt(grad[0]**2 + grad[1]**2)

def norma_razn(x_0, x_1):
    print(f"x1 = {x_1} | x0 = {x_0}")
    return sqrt((x_1[0] - x_0[0])**2 + (x_1[1] - x_0[1])**2)


def new_gradient(x0, tk):
    x_k = []
    _grad = derivF(x0[0], x0[1], k)
    
    for i in range(len(_grad)):
        _grad[i] *= tk

    for i in range(len(_grad)):
        print(f"{x0[i]} - {_grad[i]}")
        x_k.append(x0[i] - _grad[i])
        
    return x_k


k = 0
#print(f"Введите начальную точку x_{k}")
#x0 = (float(input("x: ")), float(input("y: ")))
x0 = (0.5, 1)
#print("Введите коэффициенты при x, xy, y")
#c = (float(input("c_x: ")), float(input("c_xy: ")), float(input("c_y")))
c = (2, 1, 1)
#print("Введите точность e1 и e2")
#e = (float(input("e1: ")), float(input("e2: ")))
e = (0.1, 0.15)
#print("Введите предел числа операций M")
#M = int(input("M: "))
M = 10

print(f"Введено уравнение: f(x) = {c[0]}x^2 + {c[1]}xy + {c[2]}y^2")


grad = derivF(x0[0], x0[1], k)
print(f"Градиент: {grad}")
norm = norma(x0[0], x0[1], grad, k)
print(f"Норма вектора {grad}: {norm}")

#print(f"Норма: {norma(derivF(x0[0], x0[1], k), k)}")

print(f"Шаг 4: проверка выполнения критерия окончания: {norm} < {e[0]}")
if norm < e[0]:
    print(f"Критерий выполнен, расчет закончен, x* = {grad}")
    exit

print("Критерий не выполнен. Переходим к шагу 5")
print(f"Шаг 5: проверка выполнения неравенства k >= M: {k} >= {M}")

if k >= M:
    print(f"Критерий выполнен, расчет закончен, x* = {grad}")
    exit

print("Критерий не выполнен. Переходим к шагу 6")
tk = float(input("Шаг 6. Задаем величину шага tk: "))
x_k = []

while True:
    print(f"Шаг 7. вычисляем x{k+1}")
    x_k = new_gradient(x0, tk)

    print(f"Шаг 8. Сравним f(x{k}) и f(x{k+1})")

    print(f"f(x{k}) = {f(c, x0)} | f(x{k+1}) = {f(c, x_k)}")

    if f(c, x_k) >= f(c, x0):
        print(f"Условие не выполняется для k = {k}. Зададим tk = tk/2 => {tk/2}")
        tk = tk / 2
    else:
        break

print(f"Шаг 9. Вычислим ||x1 - x0|| и |f(x1) - f(x0)|")
norm_razn = norma_razn(x0, x_k)
f_razn = abs(f(c, x_k) - f(c, x0))
print(f"norma razn: {norm_razn} < {e[0]}?\nfunc razn: {f_razn} < {e[0]}?")
#if norma_razn(x0, x_k )

```
