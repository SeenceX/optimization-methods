<h2>Численные методы многомерной оптимизации (методы первого порядка)</h2>

<h3>Метод градиентного спуска с постоянным шагом:</h3>

```python
from math import sqrt


class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'


def f(c, vector): return c[0] * vector[0] ** 2 + c[1] * vector[0] * vector[1] + c[2] * vector[1] ** 2


def deriv_f(x, y, c): return [2 * c[0] * x + c[1] * y, c[1] * x + c[2] * 2 * y]


def norma(grad): return sqrt(grad[0] ** 2 + grad[1] ** 2)


def norma_razn(x_0, x_1): return sqrt((x_1[0] - x_0[0]) ** 2 + (x_1[1] - x_0[1]) ** 2)


def new_gradient(x0, tk, c):
    xk = []
    _grad = deriv_f(x0[0], x0[1], c)

    for i in range(len(_grad)):
        _grad[i] *= tk

    for i in range(len(_grad)):
        xk.append(x0[i] - _grad[i])

    return xk


def gradient_descent_with_constant_step():
    print(f"{bcolors.OKGREEN}[Шаг 1].{bcolors.ENDC} Задаем начальные значения.")

    # print(f"Введите начальную точку x_{k}")
    # x0 = (float(input("x: ")), float(input("y: ")))
    x0 = [0.5, 1]
    # print("Введите коэффициенты при x, xy, y")
    # c = (float(input("c_x: ")), float(input("c_xy: ")), float(input("c_y")))
    c = (3, 0.1, 6)
    # print("Введите точность e1 и e2")
    # e = (float(input("e1: ")), float(input("e2: ")))
    e = (0.15, 0.2)
    # print("Введите предел числа операций M")
    # M = int(input("M: "))
    M = 10

    print(f"Введено уравнение: f(x) = {c[0]}x^2 + {c[1]}xy + {c[2]}y^2")

    k = 0
    print(f"\n{bcolors.OKGREEN}[Шаг 2]{bcolors.ENDC}. Полагаем k = {k}")
    is_done = False

    while k < M:
        grad = deriv_f(x0[0], x0[1], c)
        print(f"\n{bcolors.OKGREEN}[Шаг 3_{k}]{bcolors.ENDC}")
        print(f"Градиент: {grad}")
        print(f"Вычислим норму вектора ||{grad}||")
        norm = norma(grad)
        print(f"Норма вектора {grad}: {norm}")

        print(f"\n{bcolors.OKGREEN}[Шаг 4_{k}]{bcolors.ENDC}: проверка выполнения критерия окончания: {norm} < {e[0]}?")
        if norm < e[0]:
            print(f"{bcolors.OKGREEN}Критерий выполнен, расчет закончен, x* = {grad}{bcolors.ENDC}")
            exit()

        print(f"{bcolors.FAIL}Критерий не выполнен. Переходим к шагу 5{bcolors.ENDC}")
        print(f"\n{bcolors.OKGREEN}[Шаг 5_{k}]{bcolors.ENDC}. проверка выполнения неравенства k >= M: {k} >= {M}")

        if k >= M:
            print(f"{bcolors.OKGREEN}Критерий выполнен, расчет закончен, x* = {grad}{bcolors.ENDC}")
            exit()

        print(f"{bcolors.FAIL}Критерий не выполнен. Переходим к шагу 6{bcolors.ENDC}")
        print(f"\n{bcolors.OKGREEN}[Шаг 6_{k}].{bcolors.ENDC} Задаем величину шага tk: ", end='')
        tk = float(input())

        while True:
            print(f"\n{bcolors.OKGREEN}[Шаг 7_{k}].{bcolors.ENDC} вычисляем x{k + 1}")
            x_k = new_gradient(x0, tk, c)
            print(f"f(x{k + 1}) = {f(c, x_k)})")

            print(f"\n{bcolors.OKGREEN}[Шаг 8_{k}].{bcolors.ENDC} Сравним f(x{k}) и f(x{k + 1})")

            print(f"f(x{k}) = {f(c, x0)} | f(x{k + 1}) = {f(c, x_k)}")

            if f(c, x_k) >= f(c, x0):
                print(f"Условие не выполняется для k = {k}. Зададим tk = tk/2 => {tk / 2}")
                tk = tk / 2
            else:
                break

        print(f"\n{bcolors.OKGREEN}[Шаг 9_{k}].{bcolors.ENDC} Вычислим ||x{k + 1} - x{k}|| и |f(x{k + 1}) - f(x{k})|")
        print(f"x{k + 1} = {x_k} | x{k} = {x0}")
        norm_razn = norma_razn(x0, x_k)
        f_razn = abs(f(c, x_k) - f(c, x0))
        print(f"Норма разности: {norm_razn} < {e[1]}?\nРазность функций: {f_razn} < {e[1]}?")

        if norm_razn < e[1] and f_razn < e[1]:
            print(
                f"{bcolors.OKGREEN}Да.\nУсловия ||x{k + 1} - x{k}|| < e2 и |f(x{k + 1}) - f(x{k})| < e2 выполнены при k = {k}.\n{bcolors.ENDC}")
            if is_done:
                x0 = x_k
                print(
                    f"{bcolors.OKGREEN}Условия ||x{k + 1} - x{k}|| < e2 и |f(x{k + 1}) - f(x{k})| < e2 выполнены при k = {k - 1} и k = {k}. Расчет окончен.{bcolors.ENDC}")
                k = k + 1
                break
            else:
                is_done = True
            k = k + 1
            print(f"{bcolors.OKGREEN} k = k + 1 = {k}{bcolors.OKGREEN}")
            x0 = x_k

        else:
            if is_done:
                is_done = False
            k = k + 1
            x0 = x_k
            print(f"{bcolors.FAIL}Нет, k = k + 1 = {k}{bcolors.FAIL}")
        print('-' * 64)

    print(
        f"\n{bcolors.OKGREEN}Найдена точка{bcolors.ENDC} {bcolors.UNDERLINE}x{k} = {x0};{bcolors.ENDC} {bcolors.UNDERLINE}f(x{k}) = {f(c, x0)}{bcolors.ENDC}")


gradient_descent_with_constant_step()
```
